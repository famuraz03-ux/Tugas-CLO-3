```python
# Kode Python Soal 7 - Fibbonacci, jumlah setiap angka ada jumlah dari dua angka sebelumnya
def fib_less_than(N):
    fib_list = []
    a, b = 0, 1
    while a < N:
        fib_list.append(a)
        a, b = b, a + b
    return fib_list

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
        return True
    
def fibbonacci_primes(N):
    fib_list = fib_less_than(N)
    prime_fibs = [num for num in fib_list if is_prime(num)]
    return prime_fibs

N_value = input("Masukkan nilai N: ")
print(f"Bilangan Fibonacci kurang dari {N_value}: {fib_less_than(N_value)}")
print(f"Bilangan Fibonacci prima kurang dari {N_value}: {fibbonacci_primes(N_value)}")
```
