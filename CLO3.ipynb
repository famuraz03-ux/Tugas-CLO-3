```python
# Kode Python Soal 7 - Fibbonacci, jumlah setiap angka ada jumlah dari dua angka sebelumnya
def fib_less_than(N):
    fib_list = []
    a, b = 0, 1
    while a < N:
        fib_list.append(a)
        a, b = b, a + b
    return fib_list

def fibbonacci_primes(N):
    fibs = fib_less_than(N)
    return [x for x in fibs if x > 1 and all(x % i != 0 for i in range(2, int(x**0.5) + 1))]
    
N_value = 50
print(f"Bilangan Fibonacci kurang dari {N_value}: {fib_less_than(N_value)}")
print(f"Bilangan Fibonacci prima kurang dari {N_value}: {fibbonacci_primes(N_value)}")
```

```python
# Kode Python soal 8
def dotprod(M, i, j):
    # Ambil baris ke-i
    baris = M[i]
    
    # Ambil kolom ke-j
    kolom = [M[r][j] for r in range(len(M))]
    
    # Hitung dot product manual
    hasil = sum(baris[k] * kolom[k] for k in range(len(baris)))
    
    return hasil


# Contoh penggunaan
M = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

i = 0
j = 1

print(dotprod(M, i, j))
```

```python
# Kode Python Soal 9
import numpy as np
import matplotlib.pyplot as plt

# 1. f(x) = sqrt(|x|), x  [-9, 9]
x1 = np.linspace(-9, 9, 400)
f1 = np.sqrt(np.abs(x1))

plt.figure(figsize=(6,4))
plt.plot(x1, f1)
plt.title("f(x) = |x|")
plt.grid(True)
plt.show()


# 2. f(x) = sin(x) + sin(2x), x  [0, 4]
x2 = np.linspace(0, 4*np.pi, 400)
f2 = np.sin(x2) + np.sin(2*x2)

plt.figure(figsize=(6,4))
plt.plot(x2, f2)
plt.title("f(x) = sin(x) + sin(2x)")
plt.grid(True)
plt.show()


# 3. f(x) = 2x ln|x|, x  0, f(0)=0, x  [-1, 1]
x3 = np.linspace(-1, 1, 400)
f3 = np.zeros_like(x3)
mask = x3 != 0
f3[mask] = 2 * x3[mask] * np.log(np.abs(x3[mask]))

plt.figure(figsize=(6,4))
plt.plot(x3, f3)
plt.title("f(x) = 2x ln|x|, f(0)=0")
plt.grid(True)
plt.show()


# 4. f(x) = (x+2)(x+1)(x1)(x2)(x3), x  [-2, 3]
x4 = np.linspace(-2, 3, 400)
f4 = (x4+2)*(x4+1)*(x4-1)*(x4-2)*(x4-3)

plt.figure(figsize=(6,4))
plt.plot(x4, f4)
plt.title("f(x) = (x+2)(x+1)(x1)(x2)(x3)")
plt.grid(True)
plt.show()


# 5. f(x) = e^(x), x  [-2, 2]
x5 = np.linspace(-2, 2, 400)
f5 = np.exp(-x5**2)

plt.figure(figsize=(6,4))
plt.plot(x5, f5)
plt.title("f(x) = e^(x)")
plt.grid(True)
plt.show()
```
